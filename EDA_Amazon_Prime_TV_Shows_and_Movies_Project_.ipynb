## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


#Shape of the dataset

merged_df.shape

# So, after combining both the dataset Total number of rows = 125354 and columns = 19
````

````python
#Checking missing values for each column
merged_df.isnull().sum()
````

````python
# checking the statistics of the merged dataset using describe function and transposing it.

merged_df.describe().T

# If std =0 we call that as constant column ,then we drop such columns
````

````python
# As we can see that the columns age_certication and seasons have maximum missing values so we drop these columns

# age_certification = 68497 and seasons = 116853

merged_df.drop(["age_certification","seasons"], axis = 1, inplace = True)
````

````python
merged_df # checking if the above drop was successfull
````

````python
# Now we need to fill in the columns which have missing values but not significantly.

"""
Imputation (Replacing Values):

* Mean/Median: For numerical data.
* Mode: For categorical data.

"""

# Filter categorical columns with missing values
missing_cats = merged_df.select_dtypes(include='object').isnull().sum()
missing_cats = missing_cats[missing_cats > 0]
missing_cats
````

````python
# Now, we find the mode value of all categorical columns to replace the missing rows value

merged_df[['description', 'imdb_id','name', 'character', 'role']].mode()



""" So mode values are :
imdb_id: tt6802400
name: George 'Gabby' Hayes
character: Himself
role: ACTOR

"""
````

````python
# Filling the missing values of the cat columns with the mode values figures above

#creating a variable for the mode value description column since the string is too long

default_des = """The Three Stooges were an American vaudeville and comedy team active from 1922 until 1970,
best known for their 190 short subject films by Columbia Pictures that have been regularly airing on
television since 1958. Their hallmark was physical farce and slapstick. In films, the stooges were
commonly known by their actual first names. There were a total of
six stooges over the act's run (with only three active at any given time),
but Moe Howard and Larry Fine were the mainstays
throughout the ensemble's nearly fifty-year run."""



merged_df.fillna({
    'description': default_des,
    'imdb_id': 'tt6802400',
    'name': "George 'Gabby' Hayes",
    'character': "ACTOR",
    'role': "ACTOR"
}, inplace=True)


#Now, check if there is still any null values is the categorical columns.

merged_df.select_dtypes(include='object').isnull().sum()
````

````python
# Now, let's handle the missing values in the numerical columns

#checking which are the numerical columns with missing values that needs to replaced

missing_num = merged_df.select_dtypes(include='number').isnull().sum()
missing_num = missing_num[missing_num> 0]
missing_num
````

````python
# Now need to check the skewness of the numerical columns to see if we have to use mean or median

A=['imdb_score', 'imdb_votes',
       'tmdb_popularity', 'tmdb_score', 'person_id']
for x in A:
  print(x, '-----',merged_df[x].skew())


  # clearly skewness is not within [-0.5,0.5] that means data is skewed.
  # So lets replace missing values with median
````

````python
merged_df.loc[:,'imdb_score'].fillna(merged_df.loc[:,'imdb_score'].median(),inplace=True)
merged_df.loc[:,'imdb_votes'].fillna(merged_df.loc[:,'imdb_votes'].median(),inplace=True)
merged_df.loc[:,'tmdb_popularity'].fillna(merged_df.loc[:,'tmdb_popularity'].median(),inplace=True)
merged_df.loc[:,'tmdb_score'].fillna(merged_df.loc[:,'tmdb_score'].median(),inplace=True)
merged_df.loc[:,'person_id'].fillna(merged_df.loc[:,'person_id'].median(),inplace=True)


#After replacing with the median values now let's check if there is still any missing values.

merged_df.select_dtypes(include='number').isnull().sum()
````

````python
# As we saw that the dataset had 168 duplicate rows, let's also drop those rows.

merged_df.drop_duplicates(inplace=True)


print(merged_df.duplicated().sum()) #As we can see that there aren't any duplicate values anymore.
````

````python
#Now, lets check the final shape of the dataset after the final shape after the Data Wrangling is completed

merged_df.shape
````
